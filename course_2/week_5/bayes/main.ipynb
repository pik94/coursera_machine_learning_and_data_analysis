```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.datasets import load_breast_cancer 
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.model_selection import cross_val_score
%matplotlib inline

digits = load_digits()
breasts = load_breast_cancer()

X_dig = digits.data
y_dig = digits.target

X_brs = breasts.data
y_brs = breasts.target
```

```python
pd.concat((pd.DataFrame(X_dig), pd.DataFrame(y_dig)), axis=1).head()
```

```python
pd.concat((pd.DataFrame(X_brs), pd.DataFrame(y_brs)), axis=1).head()
```

```python
bnb = BernoulliNB()
mnb = MultinomialNB()
gnb = GaussianNB()

scores_digits = []
scores_breasts = []
```

Обучим модели на первом датасете.

```python
bnb.fit(X_dig, y_dig)
mnb.fit(X_dig, y_dig)
gnb.fit(X_dig, y_dig)
```

```python
score = cross_val_score(bnb, X_dig, y_dig).mean()
scores_digits.append(score)
score
```

```python
score = cross_val_score(mnb, X_dig, y_dig).mean()
scores_digits.append(score)
score
```

```python
score = cross_val_score(gnb, X_dig, y_dig).mean()
scores_digits.append(score)
score
```

Обучим модели на втором датасете.

```python
bnb.fit(X_brs, y_brs)
mnb.fit(X_brs, y_brs)
gnb.fit(X_brs, y_brs)
```

```python
score = cross_val_score(bnb, X_brs, y_brs).mean()
scores_breasts.append(score)
score
```

```python
score = cross_val_score(mnb, X_brs, y_brs).mean()
scores_breasts.append(score)
score
```

```python
score = cross_val_score(gnb, X_brs, y_brs).mean()
scores_breasts.append(score)
score
```
